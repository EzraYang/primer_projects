## advanced layouts: gridplot, tab and panel

```python
    # grid layout
from bokeh.layouts import gridplot

p1 = figure(plot_width=200, plot_height=200, 
            x_axis_label='fertility (children per woman)', 
            y_axis_label='female_literacy (% population)', 
            title='Asia')
p1.circle(lb.loc[lb.continent=='ASI', "population"], 
          lb.loc[lb.continent=='ASI', "female_literacy"])

p2 = figure(plot_width=200, plot_height=200, 
            x_axis_label='fertility (children per woman)', 
            y_axis_label='female_literacy (% population)', 
            title='North America')
p2.circle(lb.loc[lb.continent=='NAM', "population"], 
          lb.loc[lb.continent=='NAM', "female_literacy"])

p3 = figure(plot_width=200, plot_height=200, 
            x_axis_label='fertility (children per woman)', 
            y_axis_label='female_literacy (% population)', 
            title='Latin America')
p3.circle(lb.loc[lb.continent=='LAT', "population"], 
          lb.loc[lb.continent=='LAT', "female_literacy"])

p4 = figure(plot_width=200, plot_height=200, 
            x_axis_label='fertility (children per woman)', 
            y_axis_label='female_literacy (% population)', 
            title='Europe')
p4.circle(lb.loc[lb.continent=='EUR', "population"], 
          lb.loc[lb.continent=='EUR', "female_literacy"])

layout = gridplot( [p1, p2], 
                   [p3, p4]
                 )
output_notebook()
show(layout)
```

```python
    # panel and tab layout
from bokeh.models.widgets import Panel
from bokeh.models.widgets import Tabs

tab1 = Panel(child=p1, title='Asia')
tab2 = Panel(child=p2, title='North America')
tab3 = Panel(child=p3, title='Latin America')
tab4 = Panel(child=p4, title='Europe')

layout = Tabs(tabs=[tab1, tab2, tab3, tab4])
output_notebook()
show(layout)
```

## embed slider

```python
import pandas as pd
from bokeh.layouts import row, widgetbox
from bokeh.models import Slider
# from bokeh.charts import Histogram
from bkcharts import Histogram
from bokeh.io import show, output_notebook
from bokeh.application.handlers import FunctionHandler
from bokeh.application import Application
```

```python
    # a very simple app that updates no data
def modify_doc(doc):
    def create_figure():
        return None

    def update(attr, old, new):
        return None
    
    # widgets
    slider1 = Slider(title='slider1', start=0, end=10, step=0.1, value=2)
    slider2 = Slider(title='slider2', start=10, end=100, step=1, value=20)
    layout = widgetbox(slider1, slider2)
    doc.add_root(layout)
    
# Set up the Application 
handler = FunctionHandler(modify_doc)
app = Application(handler)

output_notebook()
show(app, notebook_url="localhost:8888")
```

```python
# Create the Document Application
def modify_doc(doc):
    def create_figure():
        p = figure(plot_width=400, plot_height=400, 
                   x_axis_label='x', y_axis_label='y')
        N = slider.value
        p.circle(random(N), random(N))
        return p

    def update(attr, old, new):
        layout.children[1] = create_figure()
        
    # widgets
    slider = Slider(title='num. of points', 
                    start=10, end=300, step=10, value=150)
    slider.on_change('value', update)
    
    # init plot
    p = create_figure()
    layout = column(widgetbox(slider, width=400), p)
    doc.add_root(layout)
    
# Set up the Application 
handler = FunctionHandler(modify_doc)
app = Application(handler)

output_notebook()
show(app, notebook_url="localhost:8888")
```

# case study: gapminder dataset

```python
data = pd.read_csv('gapminder.csv', index_col='Year')
data.info()
```

```python
data.loc[data.isnull().any(axis=1)].Country.unique()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
from sklearn.preprocessing import minmax_scale
data['scaled_popu'] = np.round(minmax_scale(data.population, feature_range=(8, 120)), 2)
```

```python
data.scaled_popu.describe()
```

```python
data.head()
```

```python
data.to_csv('gapminder_clean.csv')
```

```python
# life vs. fertility across country in 1970
sc1970 = ColumnDataSource(data={
'fertility':data.loc[1970].fertility,
'life':     data.loc[1970].life,
'country':  data.loc[1970].Country,
'region':   data.loc[1970].region,
'popu':     data.loc[1970].scaled_popu
})

mapper = CategoricalColorMapper(
        factors = ['South Asia', 'Europe & Central Asia', 
                   'Middle East & North Africa','Sub-Saharan Africa', 
                   'America', 'East Asia & Pacific'], 
        palette = ['violet', 'skyblue', 
                   'palegoldenrod', 'sandybrown', 
                   'salmon', 'lightgreen'])

p1970 = figure(plot_width=800, plot_height=400, 
       title='1970', 
       x_axis_label='Fertility', y_axis_label='Life Expectancy')
p1970.add_tools(HoverTool(tooltips='@country'))

p1970.circle(x='fertility', y='life', source=sc1970, size='popu',
     color={'field':'region', 
            'transform': mapper}, 
     legend='region')
p1970.legend.location='bottom_left'


# life vs. fertility across country in 2006
sc2006 = ColumnDataSource(data={
'fertility':data.loc[2006].fertility,
'life':     data.loc[2006].life,
'country':  data.loc[2006].Country,
'region':   data.loc[2006].region,
'popu':     data.loc[2006].scaled_popu
})

p2006 = figure(plot_width=800, plot_height=400, 
       title='2006', 
       x_axis_label='Fertility', y_axis_label='Life Expectancy')
p2006.add_tools(HoverTool(tooltips='@country'))

p2006.circle(x='fertility', y='life', source=sc2006, size='popu',
     color={'field':'region', 
            'transform': mapper}, 
     legend='region')
p2006.legend.location='bottom_left'


# link axes of two plots
layout = column(p1970, p2006)

p2006.x_range = p1970.x_range
p2006.y_range = p1970.y_range

output_notebook()
# output_file('life_vs_fertility_1970_2006.html')
show(layout)
```

```python

def modify_doc(doc):    

    def create_figure():
        year = slider.value
        sc1970 = ColumnDataSource(data={
                'fertility':data.loc[year].fertility,
                'life':     data.loc[year].life,
                'country':  data.loc[year].Country,
                'region':   data.loc[year].region,
                'popu':     data.loc[year].scaled_popu
                })
        
        mapper = CategoricalColorMapper(
                factors = ['South Asia', 'Europe & Central Asia', 
                           'Middle East & North Africa','Sub-Saharan Africa', 
                           'America', 'East Asia & Pacific'], 
                palette = ['violet', 'skyblue', 
                           'palegoldenrod', 'sandybrown', 
                           'salmon', 'lightgreen'])

        p1970 = figure(plot_width=800, plot_height=500, 
               title='Life Expectancy vs. Fertility in year {}'.format(str(year)), 
               x_axis_label='Fertility', y_axis_label='Life Expectancy')
        p1970.add_tools(HoverTool(tooltips='@country'))

        p1970.circle(x='fertility', y='life', source=sc1970, size='popu',
             color={'field':'region', 
                    'transform': mapper}, 
             legend='region')
        p1970.legend.location='bottom_left'
        return p1970

    def update(attr, old, new):
        layout.children[1] = create_figure()
        
    # widgets
    slider = Slider(title='year', start=1970, end=2006, step=1, value=1970)
    slider.on_change('value', update)
    
    # init plot
    p = create_figure()
    layout = column(widgetbox(slider, width=800), p)
    doc.add_root(layout)
    
# Set up the Application 
handler = FunctionHandler(modify_doc)
app = Application(handler)

output_notebook()
show(app, notebook_url="localhost:8888")
```
